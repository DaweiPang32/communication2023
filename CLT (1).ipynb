```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population = np.random.exponential(scale=2, size=10000)  # Non-normally distributed population
num_samples = 1000  # Number of samples to generate

# Different sample sizes to explore
sample_sizes = [1, 10, 20, 500]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Impact of Sample Size on Central Limit Theorem (Non-Normal Population)', fontsize=16)

for i, sample_size in enumerate(sample_sizes):
    # Create an array to store the sample means
    sample_means = []

    # Generate samples and calculate their means
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Plot the histogram of sample means
    row = i // 2
    col = i % 2
    axs[row, col].hist(sample_means, bins=30, density=True, alpha=0.6, color='b', label='Sample Means')
    axs[row, col].set_xlabel('Sample Mean')
    axs[row, col].set_ylabel('Frequency')
    axs[row, col].set_title(f'Sample Size {sample_size}')
    axs[row, col].axvline(np.mean(population), color='r', linestyle='dashed', linewidth=2, label='Population Mean')
    axs[row, col].legend()

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for three different populations
population_params = [
    {'distribution': 'exponential', 'scale': 2},   # Exponential distribution
    {'distribution': 'uniform', 'low': 5, 'high': 15},  # Uniform distribution
    {'distribution': 'lognormal', 'mean': 0, 'sigma': 0.5}  # Log-normal distribution
]

sample_size = 30     # Size of each sample
num_samples = 1000   # Number of samples to generate for each population

# Create an array to store the sample means for each population
sample_means = []

# Generate samples and calculate their means for each population
for params in population_params:
    distribution = params['distribution']
    if distribution == 'exponential':
        sample = np.random.exponential(scale=params['scale'], size=(num_samples, sample_size))
    elif distribution == 'uniform':
        sample = np.random.uniform(low=params['low'], high=params['high'], size=(num_samples, sample_size))
    elif distribution == 'lognormal':
        sample = np.random.lognormal(mean=params['mean'], sigma=params['sigma'], size=(num_samples, sample_size))
    
    sample_mean = np.mean(sample, axis=1)
    sample_means.append(sample_mean)

# Plot histograms of sample means for each population
plt.figure(figsize=(12, 4))
colors = ['b', 'g', 'r']
for i, params in enumerate(population_params):
    plt.subplot(1, 3, i + 1)
    plt.hist(sample_means[i], bins=30, density=True, alpha=0.6, color=colors[i], label='Sample Means')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title(f'{params["distribution"].capitalize()} Distribution')
    plt.legend()
    
plt.tight_layout()
plt.show()
```
