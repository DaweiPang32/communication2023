```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the population distributions
population_sizes = [1000, 10000, 100000]
sample_sizes = [10, 30, 50, 100]
num_samples = 1000  # Number of samples to generate for each combination

# Create subplots for different population sizes
fig, axs = plt.subplots(len(population_sizes), len(sample_sizes), figsize=(15, 10))

for i, n_pop in enumerate(population_sizes):
    for j, n_sample in enumerate(sample_sizes):
        sample_means = []

        # Generate num_samples samples of size n_sample from the population distribution
        for _ in range(num_samples):
            population = np.random.uniform(0, 1, n_pop)  # You can change the population distribution here
            sample = np.random.choice(population, n_sample)
            sample_means.append(np.mean(sample))

        # Plot the histogram of sample means
        axs[i, j].hist(sample_means, bins=30, density=True, alpha=0.6)
        axs[i, j].set_title(f'Population size = {n_pop}\nSample size = {n_sample}')
        axs[i, j].set_xlabel('Sample Mean')
        axs[i, j].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for three different populations
population_params = [
    {'distribution': 'exponential', 'scale': 2},   # Exponential distribution
    {'distribution': 'uniform', 'low': 5, 'high': 15},  # Uniform distribution
    {'distribution': 'lognormal', 'mean': 0, 'sigma': 0.5}  # Log-normal distribution
]

sample_size = 30     # Size of each sample
num_samples = 1000   # Number of samples to generate for each population

# Create an array to store the sample means for each population
sample_means = []

# Generate samples and calculate their means for each population
for params in population_params:
    distribution = params['distribution']
    if distribution == 'exponential':
        sample = np.random.exponential(scale=params['scale'], size=(num_samples, sample_size))
    elif distribution == 'uniform':
        sample = np.random.uniform(low=params['low'], high=params['high'], size=(num_samples, sample_size))
    elif distribution == 'lognormal':
        sample = np.random.lognormal(mean=params['mean'], sigma=params['sigma'], size=(num_samples, sample_size))
    
    sample_mean = np.mean(sample, axis=1)
    sample_means.append(sample_mean)

# Plot histograms of sample means for each population
plt.figure(figsize=(12, 4))
colors = ['b', 'g', 'r']
for i, params in enumerate(population_params):
    plt.subplot(1, 3, i + 1)
    plt.hist(sample_means[i], bins=30, density=True, alpha=0.6, color=colors[i], label='Sample Means')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.title(f'{params["distribution"].capitalize()} Distribution')
    plt.legend()
    
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 1, 100)
plt.hlines(y=.5,xmin=0,xmax=1,color="g")
plt.fill_between(x, 0, 0.5, where=(x <= 1), color='green', alpha=0.3)
plt.title("Uniform distribution")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Talk about normal distribution as a whole, inbetween Aryan and Dawei
# Aryan = introduce
# Obtin = Transition from Aryan to Examples 
# Dawei = Talk about sample size plots
# Arios = Talk about different population distribution plots.
```
